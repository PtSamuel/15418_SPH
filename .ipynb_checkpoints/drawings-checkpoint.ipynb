```python

# Create a black canvas
fig, ax = plt.subplots(figsize=(10, 10))
ax.set_facecolor('black')

np.random.seed(42)

# Number of circles
num_circles = 40

x_coordinates = np.random.uniform(0, 10, num_circles)
y_coordinates = np.random.uniform(0, 10, num_circles)

# Draw 20 random circles with smaller radii
for i in range(num_circles):
    x_center = x_coordinates[i]  # Random x-coordinate in [0, 1]
    y_center = y_coordinates[i]  # Random y-coordinate in [0, 1]
    radius = 0.1  # Random radius in [0.02, 0.1]
    color = (0.2, 0.4, 0.9)  # Random RGB color

    circle = plt.Circle((x_center, y_center), radius, color=color)
    ax.add_patch(circle)

# Set plot limits
ax.set_xlim(0, 10)
ax.set_ylim(0, 10)

# Display the plot
plt.show()
```

```python
# Set the random seed for reproducibility
np.random.seed(42)

# Number of points
num_points = 40

# Generate random x and y coordinates
x_coordinates = np.random.uniform(0, 10, num_points)
y_coordinates = np.random.uniform(0, 10, num_points)

# Create a 2D grid for the density map
x = np.linspace(0, 10, 1000)
y = np.linspace(0, 10, 1000)
X, Y = np.meshgrid(x, y)

# Initialize the density map
density_map = np.zeros_like(X)

# Calculate the density for each point
for i in range(num_points):
    d = np.sqrt((X - x_coordinates[i])**2 + (Y - y_coordinates[i])**2)
    density_map += np.maximum(1 - d, 0)**2

# Create a grayscale density map
plt.imshow(density_map, extent=(0, 10, 0, 10), origin='lower', cmap='viridis')

# Add colorbar
cbar = plt.colorbar()
cbar.set_label('Density')

# Set plot labels and title
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Grayscale Density Map with Cloud-Like Distribution')

# Display the plot
plt.show()
```
